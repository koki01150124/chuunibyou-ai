```python
import requests
import lxml.html
```

```python
url = "https://www.haruelf.com/entry/2019/07/19/070012"
response = requests.get(url)

response.encoding = response.apparent_encoding

print(response.text)
```

```python
html = response.text

root = lxml.html.fromstring(html)
content = root.xpath('//div[contains(@class, "entry-content")]')[0]

lines = []

# id が数字で始まる h3（1〜50）だけ対象
h3_list = content.xpath(
    './/h3[re:test(@id, "^[0-9]+")]',
    namespaces={"re": "http://exslt.org/regular-expressions"}
)

for h3 in h3_list:
    texts = []

    for elem in h3.itersiblings():
        if elem.tag == "h3":
            break
        if elem.tag != "p":
            continue

        # style付きの p は除外
        if elem.get("style"):
            continue

        text = ''.join(elem.itertext()).strip()
        if not text:
            continue

        # ================= 不要部分の除去 =================

        # 1) 広告スクリプト行は丸ごと捨てる
        if "adsbygoogle" in text:
            continue

        # 2) 「訳：〜」が入っている行は訳だとみなして捨てる
        if "訳：" in text:
            continue

        # 3) 括弧だけの注釈行を捨てる
        if (text.startswith("（") and text.endswith("）")) or \
           (text.startswith("(") and text.endswith(")")):
            continue

        # 4) 「使い方:」「【俺様系のセリフ集①】」以降は説明なので切り落とす
        cut_words = ["使い方:", "【俺様系のセリフ集①】"]
        for w in cut_words:
            if w in text:
                text = text.split(w)[0].strip()

        # 5) 記事末尾の宣伝ブロックのキーワードを削る
        trash_words = [
            "🎤セリフ練習におすすめ",
            "🎙ナレーション練習におすすめ",
            "🗣 滑舌・発声トレーニング向け",
            "🎹声劇台本（※２名以上の台本もこちら）",
            "🎧業界用語を知りたい方におすすめ",
            "🖊オーディション対策におすすめ！",
            "最後までご覧いただきありがとうございました！",
            "気になる記事があればぜひ読んでみてくださいね✨",
            "はてなブログさんのランキングにも参加しています！",
            "⇓クリックで応援お願いします！"
        ]
        for w in trash_words:
            text = text.replace(w, "")
        text = text.strip()

        # ここまでやって空になったらスキップ
        if not text:
            continue
        # ==============================================

        texts.append(text)

    lines.append(''.join(texts))

# 結果確認
# print(len(lines), "行取得")
for line in lines:
    print(line)
```

```python
output_path = "【厨二病系のセリフ集①】中二病セリフ 1〜50罰ゲームや練習用に使いください。 - コエノミカタ｜はるえるブログ.txt"

with open(output_path, mode="w", encoding="utf-8") as f:
    for line in lines:
        f.write(line + "\n")

print(f"{len(lines)} 行を書き込みました → {output_path}")
```
